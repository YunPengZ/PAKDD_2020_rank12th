```python
def get_fault(df,fault):
    t = pd.merge(df,fault,on=['serial_number','model'],how='left')
    t['tag'] = t['tag'].fillna(-1).astype(int)
    dtime = pd.to_datetime(t['dt'],format='%Y%m%d')
    fault_time= pd.to_datetime(t['fault_time'],format='%Y-%m-%d')
    faut_time_delta_days = (fault_time-dtime).dt.days
    t['get_fault_in_30_days'] = np.where(faut_time_delta_days<=30,1,0)
    return t
```

```python
def window_F1(predict,true,model_type=1,error_in_val_month=val_fault):#label==1 #更改F1的定义 在本月检测出的错误磁盘/在本月会出错的磁盘
    #删除第一列
    right_pred = predict[predict.get_fault_in_30_days==1][predict.model==2]
    nppr = len(right_pred)#预测准确的值 也只统计个数
    print('right predict:',len(right_pred))
    npp = len(predict)
    pre = nppr/npp
    print('predict:',len(predict))
    print('pre',pre)
    right_pred = get_unique_id(right_pred)
    error_in_val_month = get_unique_id(error_in_val_month)

    right_ids = error_in_val_month['unique_id'].unique()
#     error_in_val_month = error_in_val_month[error_in_val_month.model==model_type]这里应该是包括model==2的盘
#     print(right_pred[right_pred.unique_id.isin(right_ids)][['dt','fault_time']])
    ntpr = len(right_pred[right_pred.unique_id.isin(right_ids)])#当月故障的磁盘交预测正确的磁盘
    npr = len(error_in_val_month[error_in_val_month.model==2])
    recall = ntpr/npr
    print('ntpr',ntpr)
    print('npr',npr)
    print('recall:',recall)
    score = 2*pre*recall/(pre+recall)
    return score
```

```python
# use_feas
from sklearn.metrics import roc_auc_score

def get_score(val,pred,true,val_num=138,constraint=False):
    sub = val.copy()
    sub['pred'] = pred#clf.predict_proba(X_val)[:,1]
#     rate = len(val[val.get_fault_in_30_days==1])/len(val)#按错误的盘数所占的比例 test中错误的盘数大概是200个
    sub = sub[sub.model==2]
    val = val[val.model==2]
    sub = get_unique_id(sub)
    sub  = sub[sub.pred>0.020431966852624852]
#     sub = sub.sort_values('pred',ascending=False).drop_duplicates(['serial_number','model']).head(val_num)#['unique_id'].unique()
    sub = sub.sort_values('dt').drop_duplicates(['serial_number','model'])
    sub['label'] = 1
    f1_score = window_F1(sub[sub.label==1],val[val['get_fault_in_30_days']==1],error_in_val_month=val_fault)*100
    print(f1_score)
    #这样比较正常 对验证集也进行采样
    auc_score = roc_auc_score(true,pred)
    return f1_score,auc_score,sub['pred'].min()
```
